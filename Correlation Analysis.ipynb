```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
```

# Suicide data

```python
suicide_data=pd.read_csv('suicide.csv')
suicide_data
```

# Pearson Correlation Coefficient

```python
pearsn=suicide_data.corr(method='pearson')
pearsn
```

```python
sns.heatmap(pearsn,xticklabels=pearsn.columns,yticklabels=pearsn.columns,cmap='RdBu_r',annot=True)
```

# Spearman Correlation Coefficient

```python
spearman=suicide_data.corr(method='spearman')
spearman
```

```python
sns.heatmap(spearman,xticklabels=spearman.columns,yticklabels=spearman.columns,cmap='Dark2',annot=True,lw=0.5)
```

```python
sns.pairplot(suicide_data)
```

```python
r,p = scipy.stats.pearsonr(suicide_data['population'],suicide_data['suicides_no'])
```

```python
r
```

```python
p
```

# Car dataset

```python
car_dataset=pd.read_excel('car.xlsx')
car_dataset
```

```python
from pylab import rcParams
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize']=5,3
sns.set_style('ticks')
```

```python
sns.pairplot(car_dataset)
```

```python
car=car_dataset[['mpg','hp','qsec','cyl','wt']]
car
```

```python
sns.pairplot(car)
```

```python
mpg=car['mpg']
hp=car['hp']
qsec=car['qsec']
cyl=car['cyl']
wt=car['wt']
```

```python
coeff,pvalue=pearsonr(mpg,hp)
print('pearson coefficient %0.3f'%(coeff))
```

```python
pvalue
```

```python
coeff,pvalue=pearsonr(mpg,qsec)
print('pearson coefficient %0.3f'%(coeff))
```

```python
pvalue
```

```python
coeff,pvalue=pearsonr(mpg,cyl)
print('pearson coefficient %0.3f'%(coeff))
```

```python
pvalue
```

```python
coeff,pvalue=pearsonr(mpg,wt)
print('pearson coefficient %0.3f'%(coeff))
```

```python
pvalue
```
